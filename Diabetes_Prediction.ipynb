```python
# ---Check Class Imbalance (Overall)---

def imbalance_summary(y: pd.Series):
    """Return counts/percent by class, plus imbalance ratio and majority baseline."""
    counts = y.value_counts().sort_index()
    perc = (counts / counts.sum() * 100).round(2)
    maj = counts.max()
    minc = counts.min()
    imb_ratio = float(minc / maj) if maj > 0 else np.nan  # minority/majority
    maj_baseline = float(maj / counts.sum()) if counts.sum() > 0 else np.nan
    summary = pd.DataFrame({"count": counts, "percent": perc})
    return summary, imb_ratio, maj_baseline

y = cleaned_df["diabetes_status"]

overall_tbl, overall_imb, overall_maj_base = imbalance_summary(y)
print("=== Overall ====")
display(overall_tbl)
print(f"Imbalance ratio (minority/majority): {overall_imb:.4f}")
print(f"Majority-class baseline accuracy:     {overall_maj_base:.4f}")
```

```python
# ---Predictor  Outcome relationships---

from scipy import stats
from IPython.display import display

df = cleaned_df.copy()
y = df["diabetes_status"].astype(int)

def ensure_binary01(s: pd.Series) -> pd.Series:
    u = set(pd.unique(s.dropna()))
    if u.issubset({0, 1}):
        return s.astype(float)
    if u.issubset({1, 2}):
        return s.map({1: 1.0, 2: 0.0})
    raise ValueError(f"Binary expected ({s.name}); found values {sorted(list(u))}")

def outcome_rate_by_level(s: pd.Series, y: pd.Series):
    d = pd.DataFrame({"x": s, "y": y}).dropna()
    grp = d.groupby("x")["y"]
    tbl = pd.DataFrame({
        "n": grp.count(),
        "cases": grp.sum(),
        "rate": grp.mean().mul(100).round(2)
    })
    return tbl

# Binary predictors  outcome

rows = []
for col in binary_cols:
    try:
        x = ensure_binary01(df[col])  # map to {0,1}
    except Exception as e:
        print(f"[skip] {col}: {e}")
        continue

    d = pd.DataFrame({"x": x, "y": y}).dropna()
    if d.empty:
        print(f"[skip] {col}: no data after dropping NA")
        continue

    tab = pd.crosstab(d["x"], d["y"]).reindex(index=[0, 1], columns=[0, 1], fill_value=0)
    n0 = tab.loc[0].sum()
    n1 = tab.loc[1].sum()
    if n0 == 0 or n1 == 0:
        print(f"[skip] {col}: one group has zero size")
        continue

    risk0 = tab.loc[0, 1] / n0
    risk1 = tab.loc[1, 1] / n1

    rows.append({
        "predictor": col,
        "n_x0": int(n0),
        "n_x1": int(n1),
        "risk_x0": risk0,
        "risk_x1": risk1,
    })

if rows:
    out = pd.DataFrame(rows)[["predictor", "n_x0", "n_x1", "risk_x0", "risk_x1"]]
    for c in ["risk_x0", "risk_x1"]:
        out[c] = out[c].map(lambda v: f"{v*100:.2f}%")
    print("\n=== Binary predictors  outcome ===")
    print("\n Diabetes prevalence (%) by exposure  columns show n_x0/n_x1 (group sizes for x=0, x=1) and risk_x0/risk_x1 (prevalence = cases/n100 within each group)")
    display(out)

# Ordinal predictors  outcome

print("\n=== Ordinal predictors  outcome ===")
print("\n Diabetes prevalence (%) by ordered level  columns show n (level size), cases (diabetes=1), and rate (=cases/n100)")
for col in ordinal_cols:
    s = df[col]
    tbl = outcome_rate_by_level(s, y)
    print(f"\n-- {col} --")
    display(tbl)

# Nominal (multi-level) predictors  outcome

print("\n=== Nominal multi-level predictors  outcome ===")
print("\n Diabetes prevalence (%) by category  columns show n (category size), cases (diabetes=1), and rate (=cases/n100)")
for col in nominal_cols:
    s = df[col]
    tbl = outcome_rate_by_level(s, y).sort_values("rate", ascending=False)
    print(f"\n-- {col} --")
    display(tbl)
```

```python
# ---Reproducible train/test split preserving classsex proportions---

import os, random
from sklearn.model_selection import train_test_split

RNG_SEED = 42
random.seed(RNG_SEED)
np.random.seed(RNG_SEED)
os.environ["PYTHONHASHSEED"] = str(RNG_SEED)

df_analytic = cleaned_df.copy()
df_analytic["sex_label"] = df_analytic["sex"].map(sex_map)

# Stratify by joint label to preserve classsex in both splits
strat_key = df_analytic["diabetes_status"].astype(int).astype(str) + "_" + df_analytic["sex_label"]

train_raw, test_raw = train_test_split(
    df_analytic, test_size=0.20, random_state=RNG_SEED, stratify=strat_key
)

# Checks
def pct(s):
    return (s.value_counts(normalize=True).mul(100).round(2).to_dict())

print("Class % overall:", pct(df_analytic["diabetes_status"]))
print("Class % train  :", pct(train_raw["diabetes_status"]))
print("Class % test   :", pct(test_raw["diabetes_status"]))
print("Sex % overall  :", pct(df_analytic["sex_label"]))
print("Sex % train    :", pct(train_raw["sex_label"]))
print("Sex % test     :", pct(test_raw["sex_label"]))

# Save split indices for reproducibility
pd.DataFrame({"index": train_raw.index}).to_csv("train_idx.csv", index=False)
pd.DataFrame({"index": test_raw.index}).to_csv("test_idx.csv", index=False)

print("Saved split indices: train_idx.csv, test_idx.csv")
```

```python
# ---Model training, hyperparameter tuning and threshold selection---

import os, random
from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    roc_auc_score, average_precision_score, confusion_matrix, classification_report,
    precision_recall_curve, precision_recall_fscore_support
)
import xgboost as xgb

RNG_SEED = 42
random.seed(RNG_SEED); np.random.seed(RNG_SEED); os.environ["PYTHONHASHSEED"] = str(RNG_SEED)
train_idx = pd.read_csv("train_idx.csv")["index"].astype(int)
test_idx  = pd.read_csv("test_idx.csv")["index"].astype(int)

target_col   = "diabetes_status"
feature_cols = [c for c in cleaned_df.columns if c != target_col]

train_df = cleaned_df.loc[train_idx].copy()
test_df  = cleaned_df.loc[test_idx].copy()

# Make a validation split from the training pool
X_train_full = train_df[feature_cols].copy()
y_train_full = train_df[target_col].astype(int)

X_test = test_df[feature_cols].copy()
y_test = test_df[target_col].astype(int)

X_tr, X_val, y_tr, y_val = train_test_split(
    X_train_full, y_train_full,
    test_size=0.20, random_state=RNG_SEED, stratify=y_train_full
)

# Native categorical casting
if "nominal_cols" not in globals() or nominal_cols is None:
    nominal_cols = []
def cast_cats(df):
    for c in nominal_cols:
        if c in df.columns:
            df[c] = df[c].astype("category")
    return df
X_tr   = cast_cats(X_tr.copy())
X_val  = cast_cats(X_val.copy())
X_test = cast_cats(X_test.copy())

# Class weight from full training pool (natural prevailance)
n_pos_full = int((y_train_full == 1).sum()); n_neg_full = int((y_train_full == 0).sum())
scale_pos_weight = n_neg_full / max(1, n_pos_full)
print(f"[info] Train(full) prevalence -> pos={n_pos_full}, neg={n_neg_full}, scale_pos_weight={scale_pos_weight:.3f}")

# Randomized hyperparameter search
param_space = {
    "n_estimators":      [400, 600, 800, 1000],
    "max_depth":         [4, 5, 6, 8],
    "learning_rate":     [0.03, 0.05, 0.07, 0.10],
    "subsample":         [0.7, 0.8, 1.0],
    "colsample_bytree":  [0.6, 0.8, 1.0],
    "min_child_weight":  [1, 3, 5],
    "reg_alpha":         [0.0, 0.001, 0.01],
    "reg_lambda":        [1.0, 3.0, 10.0],
}
rng = np.random.default_rng(RNG_SEED + 7)
def sample_params(space): return {k: rng.choice(v) for k, v in space.items()}

base_params = dict(
    tree_method="hist",
    enable_categorical=True,
    objective="binary:logistic",
    eval_metric="aucpr",
    random_state=RNG_SEED,
    n_jobs=-1,
    scale_pos_weight=scale_pos_weight,
)

def fit_and_score(params):
    model = xgb.XGBClassifier(**{**base_params, **params})
    model.fit(
        X_tr, y_tr,
        eval_set=[(X_tr, y_tr), (X_val, y_val)],  # monitor only; no early-stopping args
        verbose=False
    )
    proba_val = model.predict_proba(X_val)[:, 1]
    pr_auc = average_precision_score(y_val, proba_val)
    return pr_auc, model

N_TRIALS = 25
best = {"score": -np.inf, "params": None, "model": None}
print("\n=== Hyperparameter search (validation PR AUC) ===")
for t in range(1, N_TRIALS + 1):
    params = sample_params(param_space)
    score, model = fit_and_score(params)
    if score > best["score"]:
        best = {"score": score, "params": params, "model": model}
    if t % 5 == 0 or t == 1:
        print(f"Trial {t:02d}/{N_TRIALS} | PR AUC={score:.4f} | params={params}")

print("\n=== Best on validation ===")
print(f"PR AUC={best['score']:.4f}")
print("Params:", best["params"])

# Retrain final model on the full training pool with best params
X_train_full_c = cast_cats(X_train_full.copy())
final_model = xgb.XGBClassifier(**{**base_params, **best["params"]})
final_model.fit(
    X_train_full_c, y_train_full,
    eval_set=[(X_tr, y_tr), (X_val, y_val)],
    verbose=False
)

# Threshold selection on validation with recall floor (maximize precision)
RECALL_FLOOR = 0.90

proba_val = final_model.predict_proba(X_val)[:, 1]
prec, rec, thr = precision_recall_curve(y_val, proba_val)  
prec_t, rec_t, thr_t = prec[1:], rec[1:], thr

mask = rec_t >= RECALL_FLOOR
if mask.any():
    idxs = np.where(mask)[0]
    best_idx = idxs[np.argmax(prec_t[idxs])]          # maximize precision subject to recall  floor
    thr_recall_floor = float(thr_t[best_idx])
    sel_p, sel_r = float(prec_t[best_idx]), float(rec_t[best_idx])
    print(f"\n[validation] Recall floor={RECALL_FLOOR:.2f} satisfied."
          f" Chosen threshold={thr_recall_floor:.4f} | P={sel_p:.3f} | R={sel_r:.3f}")
else:
    best_idx = int(np.argmax(rec_t))                  # fallback: max recall if floor unmet
    thr_recall_floor = float(thr_t[best_idx])
    sel_p, sel_r = float(prec_t[best_idx]), float(rec_t[best_idx])
    print(f"\n[validation] No point reached recall  {RECALL_FLOOR:.2f}."
          f" Using MAX-recall threshold={thr_recall_floor:.4f} | P={sel_p:.3f} | R={sel_r:.3f}")

# Save tuned model
final_model.save_model("xgb_diabetes.json")
print("\nSaved tuned model to: xgb_diabetes.json")
```

```python
# ---Performance evaluation on held-out test set - Overall and Sex-based subgroup analysis---

from sklearn.metrics import fbeta_score

# Pick the already trained model and chosen threshold
_model = model                 
THRESH = float(thr_recall_floor)         

# Build an evaluation frame for the test set
assert 'sex' in test_df.columns, "sex column not found in test_df."
proba_test = _model.predict_proba(X_test)[:, 1]
df_eval = pd.DataFrame({
    "y": y_test.values,
    "proba": proba_test,
    "sex": test_df["sex"].astype(int).values  # 0=female, 1=male
})

sex_label = {0: "female (0)", 1: "male (1)"}

def _safe_auc(y, p):
    try:
        if len(np.unique(y)) < 2:
            return np.nan
        return roc_auc_score(y, p)
    except Exception:
        return np.nan

def _safe_ap(y, p):
    try:
        return average_precision_score(y, p)
    except Exception:
        return np.nan

def _print_group_metrics(gname, gdf, thr):
    y = gdf["y"].values
    p = gdf["proba"].values
    yhat = (p >= thr).astype(int)

    # Threshold-free
    roc = _safe_auc(y, p)
    ap  = _safe_ap(y, p)
    
    # Thresholded
    P, R, F1, _ = precision_recall_fscore_support(y, yhat, average="binary", zero_division=0)
    F2 = fbeta_score(y, yhat, beta=2, average="binary", zero_division=0)
    tn, fp, fn, tp = confusion_matrix(y, yhat, labels=[0,1]).ravel()
    prev = y.mean()
    fpr = fp / (fp + tn) if (fp + tn) > 0 else np.nan
    fnr = fn / (fn + tp) if (fn + tp) > 0 else np.nan
    
    print(f"\n--- {gname} ---")
    print(f"N={len(y)} | Positives={y.sum()} ({prev:.3f} prevalence)")
    print(f"ROC AUC={roc:.3f} | PR AUC={ap:.3f}")
    print(f"Threshold={thr:.3f} | Precision={P:.3f} | Recall={R:.3f} | F1={F1:.3f}")
    print(f"                               F2={F2:.3f}")
    print(f"Confusion matrix [tn, fp, fn, tp] = [{tn}, {fp}, {fn}, {tp}]")
    print(f"FPR={fpr:.3f} | FNR={fnr:.3f}")


# Overall
_print_group_metrics("OVERALL (all test)", df_eval, THRESH)

# Subgroups by sex
for s in [0, 1]:
    gdf = df_eval[df_eval["sex"] == s]
    if gdf.empty:
        print(f"\n--- {sex_label[s]}: no rows ---")
        continue
    _print_group_metrics(f"SEX = {sex_label[s]}", gdf, THRESH)

# Fairness deltas (precision/recall gap)
def _agg_for_gap(gdf, thr):
    y = gdf["y"].values
    yhat = (gdf["proba"].values >= thr).astype(int)
    P, R, F1, _ = precision_recall_fscore_support(y, yhat, average="binary", zero_division=0)
    return P, R, F1

has_female = (df_eval["sex"] == 0).any()
has_male   = (df_eval["sex"] == 1).any()

if has_female and has_male:
    P0, R0, F10 = _agg_for_gap(df_eval[df_eval["sex"] == 0], THRESH)
    P1, R1, F11 = _agg_for_gap(df_eval[df_eval["sex"] == 1], THRESH)
    print("\n=== Precision/Recall gaps (male - female) at fixed threshold ===")
    print(f"Precision = {P1 - P0:+.3f} | Recall = {R1 - R0:+.3f} | F1 = {F11 - F10:+.3f}")

    y0 = df_eval.loc[df_eval["sex"] == 0, "y"].values 
    p0 = df_eval.loc[df_eval["sex"] == 0, "proba"].values
    y1 = df_eval.loc[df_eval["sex"] == 1, "y"].values
    p1 = df_eval.loc[df_eval["sex"] == 1, "proba"].values
    f2_0 = fbeta_score(y0, (p0 >= THRESH).astype(int), beta=2, average="binary", zero_division=0) 
    f2_1 = fbeta_score(y1, (p1 >= THRESH).astype(int), beta=2, average="binary", zero_division=0)
    print(f"F2 (male - female) = {f2_1 - f2_0:+.3f}")
    
    yhat0 = (p0 >= THRESH).astype(int)
    tn0, fp0, fn0, tp0 = confusion_matrix(y0, yhat0, labels=[0,1]).ravel()
    fpr0 = fp0 / (fp0 + tn0) if (fp0 + tn0) > 0 else np.nan
    fnr0 = fn0 / (fn0 + tp0) if (fn0 + tp0) > 0 else np.nan

    yhat1 = (p1 >= THRESH).astype(int)
    tn1, fp1, fn1, tp1 = confusion_matrix(y1, yhat1, labels=[0,1]).ravel()
    fpr1 = fp1 / (fp1 + tn1) if (fp1 + tn1) > 0 else np.nan
    fnr1 = fn1 / (fn1 + tp1) if (fn1 + tp1) > 0 else np.nan

    print(f"FPR (male - female) = {fpr1 - fpr0:+.3f} | FNR (male - female) = {fnr1 - fnr0:+.3f}")

else:
    if not has_female:
        print("\n[info] No female (sex=0) rows present in the test set.")
    if not has_male:
        print("\n[info] No male (sex=1) rows present in the test set.")
```

```python
# ---Explainability (global and local)---

import shap

_base_model = final_model
THRESH = float(thr_recall_floor)         

# Build SHAP TreeExplainer

try:
    explainer = shap.TreeExplainer(
        _base_model,
        feature_perturbation="tree_path_dependent",
        model_output="raw",
    )
    shap_ex = explainer(X_test, check_additivity=False)  # Explanation object
except Exception:
    explainer = shap.TreeExplainer(_base_model, feature_perturbation="tree_path_dependent")
    try:
        shap_vals = explainer.shap_values(X_test)
        base_val = getattr(explainer, "expected_value", 0.0)
        # Wrap into an Explanation for unified plotting
        if isinstance(shap_vals, list):
            vals = shap_vals[1] if len(shap_vals) > 1 else shap_vals[0]
        else:
            vals = shap_vals
        shap_ex = shap.Explanation(values=vals,
                                   base_values=np.full((vals.shape[0],), np.array(base_val).squeeze()),
                                   data=X_test.values,
                                   feature_names=list(X_test.columns))
    except Exception as e:
        raise RuntimeError(f"SHAP failed even with tree_path_dependent. Details: {e}")

# Convert SHAP outputs into a 2D matrix [n_samples, n_features]

def _to_matrix(expl):
    vals = getattr(expl, "values", expl)
    vals = np.asarray(vals)
    if vals.ndim == 3:
        pick = 1 if vals.shape[0] > 1 else 0
        vals = vals[pick]
    if vals.ndim == 1:
        vals = vals.reshape(-1, 1)
    return vals

shap_vals_matrix = _to_matrix(shap_ex)

# Global importance: mean |SHAP|

mean_abs = np.abs(shap_vals_matrix).mean(axis=0)
global_importance = (
    pd.DataFrame({"feature": X_test.columns, "mean_abs_shap": mean_abs})
      .sort_values("mean_abs_shap", ascending=False)
      .reset_index(drop=True)
)
print("Baseline global importance (mean|SHAP|)")
display(global_importance[["feature", "mean_abs_shap"]])
global_importance.to_csv("global_shap_importance.csv", index=False)
print("Saved: global_shap_importance.csv")

# Global SHAP bar

plt.figure()
try:
    shap.plots.bar(shap_ex, max_display=22, show=False)
    plt.title("Global SHAP")
except Exception:
    shap.summary_plot(shap_vals_matrix, X_test, plot_type="bar", max_display=22, show=False)
    plt.title("Global SHAP")
plt.tight_layout()
plt.savefig("global_shap_bar.png", dpi=200, bbox_inches="tight")
plt.show()
print("Saved: global_shap_bar.png")

# Global SHAP beeswarm

plt.figure()
try:
    shap.plots.beeswarm(shap_ex, max_display=22, show=False)
    plt.title("Global SHAP Beeswarm")
except Exception:
    shap.summary_plot(shap_vals_matrix, X_test, plot_type="dot", max_display=22, show=False)
    plt.title("Global SHAP Beeswarm")
plt.tight_layout()
plt.savefig("global_shap_beeswarm.png", dpi=200, bbox_inches="tight")
plt.show()
print("Saved: global_shap_beeswarm.png")

# Local SHAP: 3 representative cases (A: highest risk, B: near threshold, C: lowest risk)

proba_test = _model.predict_proba(X_test)[:, 1]
idx_high   = int(np.argmax(proba_test))
idx_low    = int(np.argmin(proba_test))
idx_border = int(np.argmin(np.abs(proba_test - THRESH)))

cases = [
    ("Case A  Highest predicted risk",     idx_high),
    ("Case B  Borderline around threshold", idx_border),
    ("Case C  Lowest predicted risk",      idx_low),
]

def _plot_local(idx, title):
    y_true = int(y_test.iloc[idx]) if hasattr(y_test, "iloc") else int(y_test[idx])
    p_hat  = float(proba_test[idx])
    y_hat  = int(p_hat >= THRESH)
    print(f"{title}: index={idx} | y={y_true} | p={p_hat:.3f} | ={y_hat} (thr={THRESH:.3f})")

    plt.figure()
    try:
        if hasattr(shap_ex, "__getitem__") and hasattr(shap_ex, "values"):
            single = shap_ex[idx]
            shap.plots.waterfall(single, show=False, max_display=15)
        else:
            raise TypeError("Not an Explanation object.")
    except Exception:
        vals = shap_vals_matrix[idx]
        df_local = (
            pd.DataFrame({
                "feature": X_test.columns,
                "shap": vals,
                "abs_shap": np.abs(vals),
                "value": np.asarray(X_test.iloc[idx])
            })
            .sort_values("abs_shap", ascending=False)
            .head(15)
        )
        ax = df_local.plot(kind="barh", x="feature", y="shap", legend=False)
        ax.invert_yaxis()
        plt.title(title)

    plt.title(title)
    plt.tight_layout()
    fname = f"local_{title.split('')[0].strip().replace(' ', '_').lower()}.png"
    plt.savefig(fname, dpi=200, bbox_inches="tight")
    plt.show()
    print(f"Saved: {fname}")

for tag, ix in cases:
    _plot_local(ix, tag)
```

```python
# ---Consistency of reasoning across sex---

# Mean |SHAP| per sex
male_mask   = (test_df['sex'].astype(int).values == 1)
female_mask = (test_df['sex'].astype(int).values == 0)

if male_mask.sum() == 0 or female_mask.sum() == 0:
    raise ValueError("One of the sex groups has zero rows.")

imp_male   = np.abs(shap_vals_matrix[male_mask]).mean(axis=0)
imp_female = np.abs(shap_vals_matrix[female_mask]).mean(axis=0)

df_imp = pd.DataFrame({
    "feature": X_test.columns,
    "imp_male": imp_male,
    "imp_female": imp_female
})

# Baseline global importance tables for male & female

df_male_baseline = (
    pd.DataFrame({"feature": X_test.columns, "mean_abs_shap": imp_male})
    .sort_values("mean_abs_shap", ascending=False)
    .reset_index(drop=True)
)
df_female_baseline = (
    pd.DataFrame({"feature": X_test.columns, "mean_abs_shap": imp_female})
    .sort_values("mean_abs_shap", ascending=False)
    .reset_index(drop=True)
)

# Save the subgroup baseline tables
df_male_baseline.to_csv("baseline_global_importance_male.csv", index=False)
df_female_baseline.to_csv("baseline_global_importance_female.csv", index=False)

print("\nBaseline global importance (mean|SHAP|)  MALE")
print(df_male_baseline.to_string(index=True))
print("\nSaved: baseline_global_importance_male.csv")

print("\nBaseline global importance (mean|SHAP|)  FEMALE")
print(df_female_baseline.to_string(index=True))
print("\nSaved: baseline_global_importance_female.csv")

# Rank correlation (Spearman)
r_male   = df_imp["imp_male"].rank(ascending=False, method="average")
r_female = df_imp["imp_female"].rank(ascending=False, method="average")
spearman_rho = r_male.corr(r_female, method="pearson")

# Top-K overlap (Jaccard)
def jaccard_at_k(k: int):
    k = int(min(k, len(df_imp)))
    top_m = set(df_imp.sort_values("imp_male", ascending=False).head(k).index)
    top_f = set(df_imp.sort_values("imp_female", ascending=False).head(k).index)
    inter = len(top_m & top_f)
    union = len(top_m | top_f)
    return inter / union if union > 0 else np.nan, inter, k

K_LIST = [10, 15, 20]
jac_table = []
for K in K_LIST:
    j, inter, k_eff = jaccard_at_k(K)
    jac_table.append({"K": k_eff, "Jaccard": j, "Overlap": inter})
df_jaccard = pd.DataFrame(jac_table)

# rank analysis (what moved up/down by sex)
df_imp["rank_male"]   = r_male
df_imp["rank_female"] = r_female
df_imp["delta_rank"]  = df_imp["rank_female"] - df_imp["rank_male"]   # +ve => higher (more important) in males
df_imp["delta_imp"]   = df_imp["imp_male"] - df_imp["imp_female"]

# Notable movers: largest absolute rank (filter out negligible-importance features)
notable = (
    df_imp.assign(max_imp=df_imp[["imp_male","imp_female"]].max(axis=1))
          .sort_values("delta_rank", key=lambda s: s.abs(), ascending=False)
)
N_MOVERS = 20
MIN_IMP = np.percentile(notable["max_imp"], 20)  # ignore the bottom 20% by importance
notable = notable[notable["max_imp"] >= MIN_IMP].head(N_MOVERS)

# Save artifacts
df_imp.sort_values("imp_male", ascending=False).to_csv("sex_importance_by_gender.csv", index=False)
df_jaccard.to_csv("sex_jaccard_table.csv", index=False)
notable.to_csv("sex_notable_delta_ranks.csv", index=False)

print("\nExplanation consistency by sex")
print(f"Spearman rank correlation () between male vs female mean|SHAP|: {spearman_rho:.3f}")
print("\nJaccard overlap of Top-K features (male vs female):")
print(df_jaccard.to_string(index=False))
print("\nTop movers by |rank| (saved: sex_notable_delta_ranks.csv):")
print(notable[["feature","imp_male","imp_female","rank_male","rank_female","delta_rank","delta_imp"]].to_string(index=False))

# Scatter of male vs female importance with 45 reference line
x = df_imp["imp_male"].values
y = df_imp["imp_female"].values

plt.figure(figsize=(8, 7))
plt.scatter(x, y, s=18, alpha=0.8)
mx = max(x.max(), y.max())
plt.plot([0, mx], [0, mx], linestyle="--", linewidth=1.2, color="gray")  # 45 line
plt.xlabel("Mean |SHAP|  Male")
plt.ylabel("Mean |SHAP|  Female")
plt.title("Sex-consistency  Mean |SHAP| by feature")

# Label outliers: the same "notable movers" subset (by |rank|)
labels_to_annotate = notable.copy()
for _, row in labels_to_annotate.iterrows():
    fx = row["imp_male"]
    fy = row["imp_female"]
    lbl = row["feature"]
    plt.annotate(lbl, (fx, fy), textcoords="offset points", xytext=(4, 2), fontsize=8)

plt.tight_layout()
plt.savefig("sex_consistency_scatter.png", dpi=200, bbox_inches="tight")
plt.show()
print("Saved: sex_consistency_scatter.png")
print("Saved: sex_importance_by_gender.csv, sex_jaccard_table.csv, sex_notable_delta_ranks.csv")
```

```python
# ---Explanation Stability via Bootstrap Resampling of Training Data---

from scipy.stats import spearmanr

# Config
B = 50                   # number of bootstrap refits
RANDOM_SEED = 42
TOPK_LIST = [5, 10, 15, 20]
CI_LO, CI_HI = 2.5, 97.5

rng = np.random.default_rng(RANDOM_SEED)

# Compute mean|SHAP| on the fixed test set for a fitted model
def mean_abs_shap_on_test(fitted_model, X_ref):
    # Same SHAP configuration used for baseline
    try:
        expl = shap.TreeExplainer(fitted_model,
                                  feature_perturbation="tree_path_dependent",
                                  model_output="raw")
        shap_ex_ref = expl(X_ref, check_additivity=False)
        vals = shap_ex_ref.values
    except Exception:
        # Fallback path
        expl = shap.TreeExplainer(fitted_model, feature_perturbation="tree_path_dependent")
        shap_vals = expl.shap_values(X_ref)
        base_val = getattr(expl, "expected_value", 0.0)
        if isinstance(shap_vals, list):
            vals = shap_vals[1] if len(shap_vals) > 1 else shap_vals[0]
        else:
            vals = shap_vals
    mean_abs = np.abs(np.array(vals)).mean(axis=0)
    return mean_abs

# Baseline (final_model) importance on the same fixed test set (X_test)
baseline_imp_vec = mean_abs_shap_on_test(final_model, X_test)
feature_names = list(X_test.columns)
p = len(feature_names)

df_baseline_global = (
    pd.DataFrame({"feature": feature_names, "mean_abs_shap": baseline_imp_vec})
      .sort_values("mean_abs_shap", ascending=False)
      .reset_index(drop=True)
)
df_baseline_global.to_csv("stability_baseline_global_importance.csv", index=False)
print("Saved: stability_baseline_global_importance.csv")

# Bootstrap refits: resample the training pool, refit, compute mean|SHAP| on X_test
all_imps = np.zeros((B, p), dtype=float)

for b in range(B):
    # sample with replacement from training rows
    idx = rng.integers(low=0, high=len(X_train_full), size=len(X_train_full), endpoint=False)
    Xb = X_train_full.iloc[idx].copy()
    yb = y_train_full.iloc[idx].copy()
    # Keep native categorical handling consistent
    Xb_c = cast_cats(Xb)
    # Same hyperparameters as the chosen model
    model_b = xgb.XGBClassifier(**{**base_params, **best["params"]})
    model_b.fit(Xb_c, yb, verbose=False)
    # Compute global SHAP importance on the fixed test set
    all_imps[b, :] = mean_abs_shap_on_test(model_b, X_test)

# Save bootstrap importances (rows=runs, cols=features in the same order as X_test.columns)
df_all_imps = pd.DataFrame(all_imps, columns=feature_names)
df_all_imps.to_csv("stability_bootstrap_importances_matrix.csv", index=False)
print("Saved: stability_bootstrap_importances_matrix.csv")

# Stability summaries: per-feature CI, top-k overlaps vs baseline, rank correlations vs baseline

# Per-feature distribution summaries
imp_mean   = df_all_imps.mean(axis=0).values
imp_std    = df_all_imps.std(axis=0, ddof=1).values
imp_lo     = df_all_imps.quantile(CI_LO/100.0, axis=0).values
imp_hi     = df_all_imps.quantile(CI_HI/100.0, axis=0).values

# Frequency a feature appears in top-10 among bootstrap runs
def topk_frequency(df, k):
    # count appearances in top-k (by mean|SHAP| per run)
    counts = pd.Series(0, index=df.columns, dtype=int)
    for _, row in df.iterrows():
        topk_idx = np.argsort(row.values)[::-1][:k]
        top_feats = df.columns[topk_idx]
        counts.loc[top_feats] += 1
    return counts

top10_counts = topk_frequency(df_all_imps, 10).values
top10_freq = top10_counts / B

df_feature_stability = pd.DataFrame({
    "feature": feature_names,
    "boot_mean_abs_shap": imp_mean,
    "boot_std": imp_std,
    f"boot_ci{CI_LO:.0f}": imp_lo,
    f"boot_ci{CI_HI:.0f}": imp_hi,
    "top10_freq": top10_freq
}).sort_values("boot_mean_abs_shap", ascending=False).reset_index(drop=True)

df_feature_stability.to_csv("stability_feature_summary.csv", index=False)
print("Saved: stability_feature_summary.csv")

# Rank-based stability vs. baseline (final_model)

# baseline ranks (1 = most important)
base_ranks = pd.Series(baseline_imp_vec, index=feature_names).rank(ascending=False, method="average")

spearman_rhos = []
jaccard_by_k = {K: [] for K in TOPK_LIST}

base_top_by_k = {K: set(base_ranks.sort_values().index[:0]) for K in TOPK_LIST}  # placeholder to keep keys
# Compute baseline top-K sets:
for K in TOPK_LIST:
    base_top_by_k[K] = set(pd.Series(baseline_imp_vec, index=feature_names)
                                .sort_values(ascending=False)
                                .index[:K])

for b in range(B):
    # ranks for run b
    imp_b = df_all_imps.iloc[b].values
    ranks_b = pd.Series(imp_b, index=feature_names).rank(ascending=False, method="average")

    # Spearman rho (rank correlation) vs baseline
    rho, _ = spearmanr(base_ranks.values, ranks_b.values)
    spearman_rhos.append(rho)

    # Jaccard overlap of top-K sets vs baseline
    for K in TOPK_LIST:
        topb = set(pd.Series(imp_b, index=feature_names).sort_values(ascending=False).index[:K])
        inter = len(topb & base_top_by_k[K])
        union = len(topb | base_top_by_k[K])
        j = inter / union if union > 0 else np.nan
        jaccard_by_k[K].append(j)

spearman_rhos = np.array(spearman_rhos, dtype=float)

# CI tables for Spearman and Jaccard
spearman_summary = {
    "mean_rho": float(np.nanmean(spearman_rhos)),
    f"ci{CI_LO:.0f}": float(np.nanpercentile(spearman_rhos, CI_LO)),
    f"ci{CI_HI:.0f}": float(np.nanpercentile(spearman_rhos, CI_HI)),
    "n": int(np.isfinite(spearman_rhos).sum())
}
df_jaccard_summary = pd.DataFrame({
    "K": TOPK_LIST,
    "mean": [float(np.nanmean(jaccard_by_k[K])) for K in TOPK_LIST],
    f"ci{CI_LO:.0f}": [float(np.nanpercentile(jaccard_by_k[K], CI_LO)) for K in TOPK_LIST],
    f"ci{CI_HI:.0f}": [float(np.nanpercentile(jaccard_by_k[K], HI)) if (HI := CI_HI) else None for K in TOPK_LIST],
    "n": [int(np.isfinite(jaccard_by_k[K]).sum()) for K in TOPK_LIST],
})

pd.DataFrame([spearman_summary]).to_csv("stability_rankcorr_summary.csv", index=False)
df_jaccard_summary.to_csv("stability_topk_jaccard_summary.csv", index=False)
print("Saved: stability_rankcorr_summary.csv, stability_topk_jaccard_summary.csv")

# Print summaries
print("\n=== Rank correlation stability (bootstrap vs. baseline) ===")
print(pd.DataFrame([spearman_summary]).to_string(index=False))

print("\n=== Top-K Jaccard overlap vs. baseline (distribution over runs) ===")
print(df_jaccard_summary.to_string(index=False))

# Visual artifacts
plt.figure(figsize=(7,5))
plt.hist(spearman_rhos[~np.isnan(spearman_rhos)], bins=15, alpha=0.9)
plt.xlabel("Spearman  (bootstrap run vs. baseline)")
plt.ylabel("Count")
plt.title("Explanation Stability  Rank Correlation Distribution")
plt.tight_layout()
plt.savefig("stability_spearman_hist.png", dpi=200, bbox_inches="tight")
plt.show()
print("Saved: stability_spearman_hist.png")

plt.figure(figsize=(7,5))
data = [np.array(jaccard_by_k[K], dtype=float) for K in TOPK_LIST]
plt.boxplot(data, labels=[f"Top-{K}" for K in TOPK_LIST], showmeans=True)
plt.ylabel("Jaccard overlap vs. baseline")
plt.title("Explanation Stability  Top-K Overlap")
plt.tight_layout()
plt.savefig("stability_topk_jaccard_box.png", dpi=200, bbox_inches="tight")
plt.show()
print("Saved: stability_topk_jaccard_box.png")

# Bar of top-10 frequency across bootstrap runs
df_top10_freq = df_feature_stability.sort_values("top10_freq", ascending=False).head(20)
plt.figure(figsize=(9,6))
plt.barh(df_top10_freq["feature"], df_top10_freq["top10_freq"])
plt.gca().invert_yaxis()
plt.xlabel("Frequency in Top-10 (across bootstrap runs)")
plt.title("Top-10 Feature Stability")
plt.tight_layout()
plt.savefig("stability_top10_frequency_bar.png", dpi=200, bbox_inches="tight")
plt.show()
print("Saved: stability_top10_frequency_bar.png")

# Saved files
print("\nArtifacts saved:")
print("- stability_baseline_global_importance.csv")
print("- stability_bootstrap_importances_matrix.csv")
print("- stability_feature_summary.csv")
print("- stability_rankcorr_summary.csv")
print("- stability_topk_jaccard_summary.csv")
print("- stability_spearman_hist.png")
print("- stability_topk_jaccard_box.png")
print("- stability_top10_frequency_bar.png")
```
